```python
import pandas as pd
anime_df = pd.read_csv("/content/drive/My Drive/anime_data/anime.csv")
anime_df.head()
```

```python
anime_df.shape
```

```python
anime_df["id"] = anime_df.index
anime_df.head()
```

```python
anime_df.shape
```

```python
anime_feature_df = pd.DataFrame(anime_df, columns = ['name', 'genre', 'type'])
anime_feature_df.head
```

```python
def get_important_feature(data):
  important_features = []
  for i in range(0, data.shape[0]):
    important_features.append(str(data['name'][i])+' '+str(data['genre'][i])+' '+str(data['type'][i]))
  return important_features
```

```python
anime_df['important_features'] = get_important_feature(anime_df)
anime_df
```

```python
anime_df['important_features'][0]
```

```python
#Convert the text to a matrix of token counts

from sklearn.feature_extraction.text import CountVectorizer
cm = CountVectorizer().fit_transform(anime_df['important_features'])
```

```python
#Get the consine similarity matrix from the count matrix

from sklearn.metrics.pairwise import cosine_similarity
cs = cosine_similarity(cm)
print(cs)
```

```python
#Get the shape of the cosine smilarity matrix
cs.shape
```

```python
anime_watching_name = 'Kimi no Na wa.'

#Find the anime id
anime_id = anime_df[anime_df.name == anime_watching_name]['id'].values[0]
```

```python
#Create a list of enumerations for the similarity score[(anime_id, similarity score), (...), ...]

scores = list(enumerate(cs[anime_id]))
print(scores)
```

```python
#Sort the list
sorted_scores = sorted(scores, key = lambda x:x[1], reverse = True)
print(sorted_scores)
new_sorted_scores = sorted_scores[1:]
print(new_sorted_scores)
```

```python
#Create a loop to print the first 10 similar animes

j = 0
top10_list = []
top10_cs_list = []
print('The 10 most recommended anime to', anime_watching_name, 'are:\n')
print("              name               consine similarity" )
for item in new_sorted_scores:
  anime_recommended = anime_df[anime_df.id == item[0]]['name'].values
  
  print(j+1, item[0], "\t", anime_recommended, item[1])
  top10_cs_list.append(item[1])
  top10_list.append(anime_feature_df.loc[anime_df['id'] == item[0]])
  
  j += 1
  if j > 9:
    break
```

```python
anime_feature_df[anime_df.name == anime_watching_name]
```

```python
print(top10_cs_list)
```

```python
print('The 10 most recommended anime to', anime_watching_name, 'are:\n')

top10_df = pd.concat(top10_list)
top10_df['Cosine similarity'] = top10_cs_list

top10_df
```
